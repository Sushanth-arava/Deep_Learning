### Split the datset for testing and training

```python
X_train,X_test,y_train,y_test= train_test_split(df.iloc[:,1:],df.iloc[:,0],test_size=0.2)
```

### Scale the dataset

#### Encode the y_train and y_test using label encoder

```python
X_train_tensor=torch.from_numpy(X_train_scaled)
X_test_tensor=torch.from_numpy(X_test_scaled)
y_train_tensor=torch.from_numpy(y_train)
y_test_tensor=torch.from_numpy(y_test)
```

```python
y_train_tensor.shape
```

## Lets define NN

##### For now we are using only one neuron to take in all the inputs

```python
class simpleNeuralNetwork():

  def __init__(self,X):

    self.weights=torch.rand(X.shape[1],1,dtype=torch.float64,requires_grad=True) # create a matrix for weights with the shape of first row of X and set them as 1
    self.bias=torch.zeros(1,dtype=torch.float64,requires_grad=True) #since one neuron,one bias only

  def forward(self,X):
    return torch.sigmoid(torch.matmul(X,self.weights)+self.bias)

  def calculate_loss(self,y_pred,y):
    epsilon = 1e-7
    y_pred = torch.clamp(y_pred, epsilon, 1 - epsilon)

    # Calculate loss
    loss = -(y_train_tensor * torch.log(y_pred) + (1 - y_train_tensor) * torch.log(1 - y_pred)).mean()
    return loss




```

```python
#Important parameters

learning_rate=0.1
epochs=25
```

### Create Model

```python
model=simpleNeuralNetwork(X_train_tensor)
model.weights

#define loop
for epoch in range(epochs):
  #forward pass: wx+b=z and sigmoid(z)
  y_pred=model.forward(X_train_tensor)


  #calculate loss

  loss=model.calculate_loss(y_pred,y_train_tensor)


  #calculate gradients
  loss.backward()

  #update weights
  with torch.no_grad():
    model.weights-=learning_rate*model.weights.grad
    model.bias-=learning_rate*model.bias.grad

  #zero gradients
  model.weights.grad.zero_()
  model.bias.grad.zero_()


  print(f'Epoch: {epoch+1}, Loss: {loss.item()}')



```

### model evaluation

```python
with torch.no_grad():
  y_pred = model.forward(X_test_tensor)
  y_pred = (y_pred > 0.5).float()
  print(y_pred)
  accuracy = (y_pred == y_test_tensor).float().mean()
  print(f'Accuracy: {accuracy.item()}')
```
